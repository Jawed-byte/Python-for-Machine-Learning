Q. Find out the batsman who hit highest fours?

```python
mask = delivery['batsman_runs']==4
new_delivery = delivery[mask]
```

```python
# Total no of fours hit in IPL
new_delivery.shape[0]
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

```python
mask1 = delivery['batsman_runs']==6
new_delivery1 = delivery[mask1]
```

```python
# Total no of sixes hit in IPL
new_delivery1.shape[0]
```

```python
new_delivery1.groupby('batsman')['batsman_runs'].count().sort_values(ascending=False).head(10)
```

Q. Which are the 3 teams against which Virat Kohli has scored highest runs?

```python
vk = delivery[delivery['batsman']=='V Kohli']
```

```python
vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
def run_scored(batsman_name):
  vk = delivery[delivery['batsman']==batsman_name]
  return vk.groupby('bowling_team')['batsman_runs'].sum().sort_values(ascending=False).head(1).index[0]
```

```python
run_scored('V Kohli')
```

```python
run_scored('DA Warner')
```

```python
run_scored('G Gambhir')
```

```python
run_scored('MS Dhoni')
```

Q. Find the most destructive death over batsman in the history of IPL

```python
# isin()
# Strike Rate = (Number of runs / Number of balls)/100
# Min batsman 200 balls in over 16-20
```

```python
death_over = delivery[delivery['over'] > 15]
```

```python
death_over.groupby('batsman')['batsman_runs'].count()
```

```python
death_over.groupby('batsman')['batsman_runs'].count() > 200
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
x = all_batsman>200
all_batsman[x]
```

```python
batsman_list = all_batsman[x].index.tolist()
```

```python
# Runs scored by all these 43 batsman
# Balls played by these 43 batsman

final = delivery[delivery['batsman'].isin(batsman_list)]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
```

```python
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```
